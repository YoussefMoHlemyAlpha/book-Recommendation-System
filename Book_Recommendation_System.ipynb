## Feature Selection

```python
data=data.drop(['Rating','UPC','Product Type','Price (excl. tax)','Price (incl. tax)','Tax','Availability','Number of reviews'],axis=1)
```

```python
data.head()
```

```python
data['details']=data['Category']+data['Description']
```

```python
data=data.drop(['Category','Description'],axis=1)
```

```python
data.head()
```

```python
cv=CountVectorizer(max_features=1000,stop_words='english')
vector=cv.fit_transform(data['details'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommand(book):
  index=data[data['Title']==book].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True , key=lambda vector:vector[1])
  for i in distance[1:6]:
    print(data.iloc[i[0]].Title)
```

```python
recommand('Sapiens: A Brief History of Humankind')
```
